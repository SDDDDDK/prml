# ヒストグラム密度推定

## 問題：確率密度関数(pdf)の推定
混合ガウス分布：

$$
p(x) = 0.7 \mathcal{N}(x \mid 0, 1) + 0.3 \mathcal{N}(x \mid 4, 1)
$$

```julia
using Plots
using Statistics
using Random

Random.seed!(3)

function gauss(x; μ=0.0, σ=1.0)
    exp(-(x-μ)^2/(2σ^2)) / (sqrt(2pi)σ)
end

function genrandom(n)
    r = randn(n)
    mask = rand(n) .< 0.3
    r[mask] .= (randn(n) .+ 4.0)[mask]
    r
end

left = -4
right = 8
xline = collect(range(left, right, length=10000))
plot(xline, 0.7 .* gauss.(xline) .+ 0.3 .* gauss.(xline, μ=4.0, σ=1.0), label="true pdf", color="red")
```

```julia
N = 100
Δ = 0.1

histogram(genrandom(N), bins= floor(Integer, (right - left) / Δ), weights=ones(N)./(N*Δ), label="histogram estimation", color="lightblue")
plot!(xline, 0.7 .* gauss.(xline) .+ 0.3 .* gauss.(xline, μ=4.0, σ=1.0), label="true pdf", color="red")
```

```julia
Δ = 1

histogram(genrandom(N), bins= floor(Integer, (right - left) / Δ), weights=ones(N)./(N*Δ), label="histogram estimation", color="lightblue")
plot!(xline, 0.7 .* gauss.(xline) .+ 0.3 .* gauss.(xline, μ=4.0, σ=1.0), label="true pdf", color="red")
```

```julia
Δ = 2

histogram(genrandom(N), bins= floor(Integer, (right - left) / Δ), weights=ones(N)./(N*Δ), label="histogram estimation", color="lightblue")
plot!(xline, 0.7 .* gauss.(xline) .+ 0.3 .* gauss.(xline, μ=4.0, σ=1.0), label="true pdf", color="red")
```
